```python
from datasets import load_dataset

dataset = load_dataset("simecek/czech_news")
new_dataset = []


## 0 - je concatenne headlines, brief et content avec les poids
for i in range(len(dataset["train"])):
    article = dataset['train'][i]
    full = article["headline"] + article["headline"] + article["brief"] + article["brief"] + article["content"]
    new_dataset.append(full)



### 1 - indexation
from rank_bm25 import BM25Okapi
# BM25 attend une liste de listes de MOTS
tokenized_docs = [doc.split() for doc in new_dataset]
bm25 = BM25Okapi(tokenized_docs)


### 2 - Requête
query = "volby Praha"
tokenized_query = query.split()
scores = bm25.get_scores(tokenized_query)
```
